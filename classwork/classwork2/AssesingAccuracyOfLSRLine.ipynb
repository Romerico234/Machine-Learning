```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0, 2, size=30)
print(x)
```

```python
def simulate_data(x, std):
    return -1.6*x + 0.8 + np.random.normal(0, std , size=len(x))
```

```python
slopes = []
intercepts = []

for i in range(0, 1000):
    y = simulate_data(x, 0.5)
    m_hat, b_hat = np.polyfit(x, y, 1)
    slopes.append(m_hat)
    intercepts.append(b_hat)
    print(f"m_hat = {m_hat} and b_hat = {b_hat}")
```

```python
mean_of_slopes = np.mean(slopes)
mean_of_intercepts = np.mean(intercepts)
print(f"Mean of the Slopes = {mean_of_slopes}\nMean of the Intercepts = {mean_of_intercepts}")
```
